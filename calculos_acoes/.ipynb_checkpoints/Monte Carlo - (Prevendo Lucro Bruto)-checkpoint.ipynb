```python
import numpy as np
import matplotlib.pyplot as plt

#OBJETIVO
#Prever o lucro bruto de uma empresa

#Requisitos
#-Receitas esperadas
#-CMV esperado(Custo da Mercadoria Vendida) -  EUA = COGSS
```

```python
rev_m = 170 #receitas esperadas 170m
rev_stdev = 20 #desvio padrao 20m
iterators = 1000 #numero de iterações
```

```python
rev = np.random.normal(rev_m, rev_stdev, iterators)
rev
```

```python
plt.figure(figsize=(15,6))
plt.plot(rev)
```

```python
COGS = - (rev * np.random.normal(0.6,0.1)) #CMV ~ 60% - das receitas são gastas para produzir a mercadoria aproximadamente.
#CMV = percentual das receitas
#0.6 é a media
#0.1 é o desvio padrão

plt.figure(figsize=(15, 6))
plt.plot(COGS)
```

```python
COGS.mean()
```

```python
COGS.std()
```

```python
#A diferença entre as receitas e o CMV = lucro bruto

Gross_Profit = rev + COGS
Gross_Profit

plt.figure(figsize=(15, 6))
plt.plot(Gross_Profit) 

#Resultado = Distribuição normal com a diferença entre as médias das receitas e as médias do CMV
```
