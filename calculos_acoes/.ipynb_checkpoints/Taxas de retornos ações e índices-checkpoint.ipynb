```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['FLRY3.SA', 'ITSA4.SA', 'WEGE3.SA', 'UNIP6.SA']
mydata = pd.DataFrame()
for t in tickers: 
    mydata[t] = wb.DataReader(t, data_source='yahoo', start='1995-1-1')['Adj Close']
```

# Taxa de retorno simples do Portfolio.

```python
returns = (mydata / mydata.shift(1)) - 1
returns.head()
```

```python
weights = np.array([0.25, 0.25, 0.25, 0.25]) #pesos da carteira 1
```

```python
np.dot(returns, weights) #calcula o produto de um vetor ou matriz; Resultado = Produto escalar
```

```python
annual_returns = returns.mean() * 250 #retorno médio de cada ação anual
annual_returns
```

```python
pfolio_1 = str(round(np.dot(annual_returns, weights), 5) * 100) + '%'
print(pfolio_1)
```

```python
weights_2 = np.array([0.4, 0.4, 0.15, 0.05]) #novos pesos da carteira  2
pfolio_2 = str(round(np.dot(annual_returns, weights_2), 5) * 100) + '%'
```

```python
print(pfolio_2)
```
